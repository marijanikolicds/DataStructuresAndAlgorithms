```python
class Node:
        def __init__(self, val = None, R = 26):
            self.val = val
            self.next = [None] * R
```

```python
class Trie:
    R = 26 # 256 - extended ASCII    
    root = Node()
            
    def insert(self, key, val):
        return self.insert_util(self.root, key, val, 0)
    
    def insert_util(self, node, key, val, d):
        if node is None:
            node = Node()
            
        if len(key) == d:
            node.val = val
            return node
        
        idx = ord(key[d]) - ord('a')
        node.next[idx] = self.insert_util(node.next[idx], key, val, d + 1)
        return node
    
    def get(self, key):
        node = self.get_util(self.root, key, 0)
        return node.val if node else None
    
    def get_util(self, node, key, d):
        if node is None:
            return None
        
        if len(key) == d:
            return node
        
        idx = ord(key[d]) - ord('a')
        return self.get_util(node.next[idx], key, d + 1)
        
    def keys(self):
        q = []
        self.collect(self.root, "", q)
        return q
        
    def collect(self, node, prefix, q):
        if node is None:
            return
        
        if node.val != None:
            q.append(prefix)
            
        for idx in range(self.R):
            ch = chr(idx + ord('a'))
            self.collect(node.next[idx], prefix + ch, q)
                    
    def keys_with_prefix(self, prefix):
        node = self.get_util(self.root, prefix, 0)
        q = []
        self.collect(node, prefix, q)
        return q
    
    def longest_prefix(self, query):
        l = self.search(self.root, query, d = 0, l = 0)
        return query[:l]
    
    def search(self, node, query, d, l):
        if node is None:
            return l
        
        if node.val is not None:
            l = d
            
        if len(query) == d:
            return l
        
        idx = ord(query[d]) - ord('a')
        return self.search(node.next[idx], query, d + 1, l)
```
